# Write and Run Scraping functions

```python
def SearchResults(posts):
    for i,post in enumerate(totalPosts):
        postPrice = post.find('div',class_='price').get_text()
        print(postPrice)
        postPlace = post.find('div',class_='listing__address').get_text()
        print(postPlace)
        postSQM = post.find('div',class_='listing__data--area-size').get_text()
        print(postSQM)
        postRoom = post.find('div',class_='listing__data--room-count').get_text()
        print(postRoom)
        Price.append(postPrice)
        Place.append(postPlace)
        Size.append(postSQM)
        Rooms.append(postRoom)
        
def NextPage(posts,link):
    try:
        driver = webdriver.Chrome(options=options,executable_path = r"C:\Users\Matyas\Documents\Untitled Folder\webdriver\chromedriver.exe")
        driver.get(link)
        #html = requests.get(link)
        soup = BeautifulSoup(driver.page_source,'html.parser')
        #nextButton = driver.find_element_by_xpath("//*[contains(text(), 'Következő')]")
        nextButton = soup.find('a',text ='Következő oldal')

        #print(nextButton.get_attribute('href'))
        #print(nextButton.get('href'))
        posts.extend(soup.find_all('div','listing__card'))
        print(posts)
        if nextButton is None:
            return posts
        #return NextPage(posts, nextButton.get_attribute('href'))
        else:
            return NextPage(posts,nextButton.get('href'))
    finally:
        driver.quit()
```
